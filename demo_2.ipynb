### 1. Looking at Some Pretty Faces

```python
#Initialization
face = FaceTracker(name="BlazeFace")
```

```python
#Single Example
# .resize((128,128))
img1 = Image.open("misc/hyesun.png").resize((128,128))
img2 = Image.open("misc/bts.jpg")
```

```python
img_cv1.shape
```

```python
#Infer and visualize
output = face.predict(img_cv1)
face.visualize(img_cv1,output)
```
